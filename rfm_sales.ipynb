# Customer Segmentation Analysis Using Recency, Frequency, Monetary (RFM) 
RFM Analysis is a technique to know customer behavior based on customers historical transactions where the analysis is focuses on recency, frequencey and purchase amount were made by customers

```python
# Import Pandas
import pandas as pd
```

```python
# Import the data where dataset i download from kaggle.com
df=pd.read_csv(r"C:\Users\win10\Downloads\DA\salesfolder\sales.csv")
```

```python
# Set this function to fully look all the columns 
# We can see our head of new dataset
pd.set_option('display.max_columns', None)
df.head()
```

```python
# checking the null values
df.isnull().sum()
```

```python
# In this cell we calculating the recency of a customer where the definition of recency is to see when the customer made her last order
# where the thing we can do to calculating is make a reference date to start calculating and then make the difference between the two dates to calculate the amount of time passed
df['order_date']=pd.to_datetime(df['order_date'])
df_recency=df
reference_date=pd.to_datetime('1-1-2022')
df_recency=df_recency.groupby(by='cust_id', as_index=False)['order_date'].max()
df_recency.columns=['cust_id','max_date']

df_recency['Recency']=df_recency['max_date'].apply(lambda row:(reference_date-row).days)
df_recency.drop('max_date', inplace=True, axis=1)
df_recency[['cust_id','Recency']]
```

```python
# Calculating the frequency order of each customer
df_frequency=df
df_frequency=df_frequency.groupby('cust_id')['order_id'].nunique()
df_frequency.reset_index()
df_frequency.columns=['cust_id', 'Frequency']
df_frequency.head()
```

```python
# In this cell we calculating the monetary where the monetary is how much customer spend their money for their purchases
df_monetary=df
df_monetary['Monetary']=df_monetary['qty_ordered']*df_monetary['price']
df_monetary=df_monetary.groupby('cust_id', as_index=False)['Monetary'].sum()
df_monetary.columns=['cust_id','Monetary']
df_monetary.head()
```

```python
# Merge the recency ,frequency and monetary on customer id
rf=df_recency.merge(df_monetary, how='inner', on='cust_id')
rfm=rf.merge(df_frequency, how='inner', on='cust_id')
rfm.rename({'order_id':'Frequency'}, axis=1, inplace=True)
rfm.head()
```

# Define RFM Score
Based on the RFM Table i will assign score to Recency, Frequency, Monetary (RFM) between 1-3  for each RFM value where 3 is the best score and 1 is the worst score 

```python
# This cell calculate recency score
def recency_score(x):
    if x['Recency'] <=93:
        recency = 3
    elif x['Recency'] >90 and x['Recency'] <=120:
        recency = 2
    else:
        recency = 1
    return recency
rfm["R"]=rfm.apply(recency_score, axis=1)
            
```

```python
# This cell calculate frequency score
def frequency_score(x):
    if x['Frequency'] <=4:
        frequency = 1
    elif x['Frequency'] >4 and x['Frequency'] <=8:
        frequency = 2
    else:
        frequency = 3
    return frequency
rfm["F"]=rfm.apply(frequency_score, axis=1)
```

```python
pd.qcut(rfm['Monetary'], q=3).head()
```

```python
# This cell calculate monetary score
monetary_score=pd.qcut(rfm['Monetary'], q=3, labels=range(1,4))
rfm=rfm.assign(M=monetary_score.values)
```

```python
# Combine each RFM value into one columns 'RFM_Score' to know the final score
def rfm_score(row):
    return str(row['R']) + str(row['F']) + str(row['M'])
rfm['RFM_Score']=rfm.apply(rfm_score, axis=1)
```

# RFM Segments
In this section i make clustering or segmentation based on RFM score and i divided into 7 segments where the segment is Best Customer, Potential 1, Potential 2,Need Attention 1, Need Attention 2, Lost 1, Lost 2. And the things i add number in potential,need attention, lost like 1 or 2 because i want to define the ranking where number 2 after those statement is worse than number 1


```python
# In this cell is Best Customer where best customer have the highest RFM Score
rfmgroup=rfm[rfm['RFM_Score']=='333']

rfmgroup['Segment']='Best Customer'
rfmgroup.head()
```

```python
# In this cell is to define potential 1 group
rfmgroup2=rfm[rfm['RFM_Score'].str.contains('332|331|323|313')]

rfmgroup2['Segment']='Potential_1'
rfmgroup2.head()
```

```python
# In this cell is to define potential 2 group
rfmgroup3=rfm[rfm['RFM_Score'].str.contains('321|322|311|312')]

rfmgroup3['Segment']='Potential_2'
rfmgroup3.head()
```

```python
# In this cell is to define need attention 1 group
rfmgroup4=rfm[rfm['RFM_Score'].str.contains('233')]

rfmgroup4['Segment']='Need Attention_1'
rfmgroup4.head()
```

```python
# In this cell is to define need attention 2 group
rfmgroup5=rfm[rfm['RFM_Score'].str.contains('223|213|212|231|232|211|221|222')]

rfmgroup5['Segment']='Need Attention_2'
rfmgroup5.head()
```

```python
# In this cell is to define need lost 1 group
rfmgroup6=rfm[rfm['RFM_Score'].str.contains('132|123|113|133')]

rfmgroup6['Segment']='Lost_1'
rfmgroup6.head()
```

```python
# In this cell is to define lost 2 group
rfmgroup7=rfm[rfm['RFM_Score'].str.contains('111|112|121|122|131')]

rfmgroup7['Segment']='Lost_2'
rfmgroup7.head()
```

```python
# Concat the segment group 1-7 into dataframe
rfm_data=pd.concat([rfmgroup,rfmgroup2,rfmgroup3,rfmgroup4,rfmgroup5,rfmgroup6,rfmgroup7])
```

```python
# This function is to see the total customer each segments
rfm_data['Segment'].value_counts()
```

```python
# Merge the dataframe without RFM that firstly we added with dataframe with RFM and Segment on customer ID 
data_rfm=rfm_data.merge(df, how='inner', on='cust_id')
data_rfm.head()
```
